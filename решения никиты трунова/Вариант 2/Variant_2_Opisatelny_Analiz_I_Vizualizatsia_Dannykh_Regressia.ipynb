## Импорты

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('seaborn')
```

```python
def analyze_numeric(feature):
    fig, ax = plt.subplots(1, 2, figsize=(15, 5))
    sns.histplot(df[feature], ax=ax[0])
    sns.regplot(x=feature, y=target, data=df, ax=ax[1])
    plt.show()

def analyze_categorical(feature):
    fig, ax = plt.subplots(1, 2, figsize=(15, 5))
    df[feature].value_counts().plot.pie(ax=ax[0], autopct="%.1f%%")
    sns.countplot(x=df[feature], ax=ax[1])
    plt.show()
    fig, ax = plt.subplots(1, 1, figsize=(13.5, 5))
    sns.boxplot(x=feature, y=target, data=df, ax=ax)
    plt.show()
```

### Связь переменных

```python
corr = df.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(
    corr,
    annot=True,
    vmin=-1
)
plt.show()
```

```python
threshold = 0.8
plt.figure(figsize=(15, 10))
sns.heatmap(
    corr[(corr < -threshold) | (corr > threshold)],
    annot=True,
    vmin=-1
)
plt.show()
```
